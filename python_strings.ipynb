```python
#find length of given string without using len() function

string = input('Enter a string: ')

counter = 0
for i in string:
  counter+=1
print("Length of string is:", counter)
```

```python
# Extract username from a given email.
# Eg if the email is nitish24singh@gmail.com
# then the username should be nitish24singh

string = input('Enter a string: ')

pos = string.index('@')
print(string[0:pos])
```

```python
# Count the frequency of a particular character in a provided string.
# Eg 'hello how are you' is the string, the frequency of h in this string is 2.

s = input('Enter a string: ')
term = input('Enter the character whose freq need to check: ')

counter = 0

for i in s:
  if i == term:
   counter+=1

print('frequency is: ',counter)
```

```python
# WAP which can remove a particular charater from a string

s = input('Enter a string: ')
word = input('Enter the character u wanna remove: ')

result = ''

for i in s:
  if i != word:
    result = result + i
print(result)
```

```python
# Check if a string is palindrome or not

s = input('Enter the string: ')

flag = True

for i in range(0, len(s) // 2):
    if s[i] != s[len(s) - i - 1]:
        flag = False
        print('Not a Palindrome')
        break

if flag:
    print('Palindrome')
```

```python
#WAP to count the number of words in a string without split function

s = input('Enter the string: ')

L = []
result = ''

for i in s:
  if i != ' ':
    result = result + i

  else:
    L.append(result)
    result = ''

print(L)

```

```python
# Given input string ko Titlecase mai convert karo without using title()

s = input('Enter the string: ')
L = []

for i in s.split():
  L.append(i[0].upper() + i[1:].lower())

print(" ".join(L))
```
